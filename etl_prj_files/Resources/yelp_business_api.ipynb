```python
#Dependencies 

import pandas as pd 
import requests
from pprint import pprint
from config import api_key
import json
import numpy as np
```

```python
#File path
datapath6 = "one-star-michelin-restaurants.csv"
datapath7 = "two-stars-michelin-restaurants.csv"
datapath8 = "three-stars-michelin-restaurants.csv"
datapath9 = "lat_long.csv"
```

```python
# Read in CSVs
michelin_one = pd.read_csv(datapath6)
michelin_two = pd.read_csv(datapath7)
michelin_three = pd.read_csv(datapath8)
```

```python
#Concattinating all three CSVs
michelin_data = pd.concat([michelin_one, michelin_two, michelin_three])
michelin_data = michelin_data.reset_index(drop=True)
michelin_data
```

```python
#CLEANING DATA : Fill NaN Values 
values = {'city': 'Hong Kong', 'price': " "}
michelin_data = michelin_data.fillna(value=values)
```

```python
michelin_data.head()
```

```python
#Define URL
url = 'https://api.yelp.com/v3/businesses/search'
```

```python
#Define Params and Headers
headers = {'Authorization': 'Bearer %s' % api_key}
```

```python
#Loop through Michelin_Data to find match with Yelp Business.Search API
response_list = []
for i in range(len(michelin_data)):
    params = {"term":"", "location": "", "latitude": "", "longitude":""}
    restaurant = michelin_data.iloc[i]
    params['term'] = restaurant['name']
    params['location'] = restaurant['city'] + ", " + restaurant['region']
    params['latitude'] = float("{0:.5f}".format(restaurant['latitude']))
    params['longitude'] = float("{0:.5f}".format(restaurant['longitude']))
    print(f"{params['term']}: ")
    try:
        response = requests.get(url, params=params, headers=headers)
        response_list.append(response)

        print(response)
    except:
        print('hi')
        break
```

```python
# Create DataFrame to insert values
df = pd.DataFrame(columns=['business_id', 'business_name', 'latitude', 'longitude', 'review_count', 'rating'])

```

```python
#Create a for loop to match Michelin_data with Yelp API requests and append each row to data frame
class_list = []
i = 0
for r in response_list: 
    
        try:
                rjson = r.json()
                #print(rjson['businesses'])
                if rjson['businesses'] != []:
                    #print(rjson['businesses'])
                    
                    
               
                
                    q =[rjson['businesses'][0]['id'],
                                   rjson['businesses'][0]['name'],
                                   rjson['region']['center']['latitude'],
                                   rjson['region']['center']['longitude'],
                                   rjson['businesses'][0]['review_count'],
                                   rjson['businesses'][0]['rating']]
                    df.loc[i] = q
                    i+= 1
#                     print("TEST HERE", q)
                
                

        except IndexError:
            print('Not business found')

        except KeyError:
            print('No Key found')
            

   
            
            
```
